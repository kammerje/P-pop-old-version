```python
import numpy as np
import planet_sampler_pop as ps
from matplotlib import pyplot as plt
```

```python
data = ps.PlanetSample("emile_testing_AG13_bright5_baseline.txt", dataset= "SAG")
```

```python
data.HZ_in
```

```python
plt.figure(figsize = (14,12))
plt.scatter(data.HZ_in, data.Ts, label = "HZ in",  color = "green" , s = 3)
plt.scatter(data.HZ_out, data.Ts, label = "HZ out", color = "blue", s = 3)
ax = plt.gca()
plt.title("Dynamic HZ")
plt.ylabel("Host star effective temperature [K]")
plt.xlabel("$S_{eff}$ $[S_{sun}]$")
ax.invert_xaxis()
plt.legend()
plt.savefig("Dynamic HZ.png")
```

```python
print(data.Ts[np.argmin(data.HZ_out)], min(data.HZ_out))
```
